Q10: Produce a list of facilities with a total revenue less than 1000.
The output of facility name and total revenue, sorted by revenue. Remember
that there's a different cost for guests and members! */

```python
cur.execute("SELECT name, (membercost*(SELECT SUM(slots) FROM Bookings WHERE Bookings.facid = Facilities.facid AND Bookings.memid >0)+guestcost *(SELECT SUM(slots) FROM Bookings WHERE Bookings.facid = Facilities.facid AND Bookings.memid =0)) AS revenue FROM Facilities WHERE revenue < 1000 ORDER BY revenue;")
cur.fetchall()
```

Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order 

```python
cur.execute("SELECT fullname, referrer FROM (SELECT a.surname ||' '|| a.firstname AS fullname, b.surname ||' '|| b.firstname AS referrer FROM Members as a INNER JOIN Members as b WHERE b.memid = a.recommendedby ORDER BY a.surname, a.firstname);")
cur.fetchall()
```

Q12: Find the facilities with their usage by member, but not guests

```python
cur.execute("SELECT f.name, b.slots || ' slots', m.surname||', '||m.firstname AS fullname FROM Members as m LEFT JOIN bookings as b USING(memid) LEFT JOIN Facilities as f USING(facid) WHERE b.memid <> 0 ORDER BY fullname;")
cur.fetchall()
```

Q13: Find the facilities' usage by month, but not guests

```python
cur.execute("SELECT f.name, SUM(b.slots), strftime('%m',starttime) as month, SUM(b.slots)*0.5 AS hours_booked_monthly FROM Members as m LEFT JOIN bookings as b USING(memid) LEFT JOIN Facilities as f USING(facid) WHERE b.memid <> 0 GROUP BY month, name ORDER BY name;")
cur.fetchall()
```
